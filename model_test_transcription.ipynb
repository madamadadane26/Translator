```python
import os
import numpy as np
import librosa
import tensorflow as tf
from tensorflow.keras.models import load_model
from tensorflow.keras.utils import to_categorical

# Load the trained model
model = load_model('trained_model.h5')

# Define the LibriSpeech dataset directory
dataset_dir = 'C:\\Users\\jenny\\Notebooks\\LibriSpeech\\train-clean-100\\19\\198\\testing'

# Loop through all the audio files in the directory and transcribe them
for root, dirs, files in os.walk(dataset_dir):
    for file in files:
        if file.endswith('.flac'):
            # Load the audio file using Librosa
            audio_file = os.path.join(root, file)
            audio_data, sample_rate = librosa.load(audio_file, sr=16000)

            # Split the audio file into segments of 1 second each
            segment_size = 16000 # 1 second at 16000 Hz
            segments = [audio_data[i:i+segment_size] for i in range(0, len(audio_data), segment_size)]

            # Preprocess the audio data for each segment and transcribe
            transcriptions = []
            for segment in segments:
                # Pad the segment with zeros to ensure it has the correct length
                segment = tf.keras.preprocessing.sequence.pad_sequences([segment], maxlen=38400, dtype='float32', padding='post', truncating='post')

                # Reshape the segment
                segment = segment.reshape(-1, 300, 128, 1)

                # Use the model to transcribe the segment
                transcription = model.predict(segment)
                transcription = tf.keras.backend.get_value(transcription)[0]

                # Add the transcription to the list
                transcriptions.append(transcription)

            # Combine the transcriptions and output as text
            # Combine the transcriptions and output as text
            transcription_text = ' '.join([str(tf.keras.backend.get_value(tf.argmax(t))) for t in transcriptions])
            print(f'Transcription for {audio_file}: {transcription_text}')

```
